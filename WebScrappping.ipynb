```python
url = "http://marvel-ironman.surge.sh/"
html = urlopen(url) 
soup = BeautifulSoup(html, 'lxml') 
```

```python
marv_tab=[]
marv_tab = soup.find_all("div", class_="col-lg-8 mx-auto")
```

```python
for x in marv_tab:
    print(x)
    if x == 2:
        break
```

```python
names=[]
ids=[]
aligns=[]
for x in marv_tab:
    name=soup.find("h2")
    name_text=name.get_text()
    names.append(name_text)
    id=soup.find("h4")
    id_text=id.a.get_text()
    ids.append(id_text)
    align=soup.find("h4")
    align_text=align.a.get_text()
    aligns.append(align_text)
```

```python
df = pd.DataFrame({'Character Name':names,'Identity':ids,'Align':aligns}) 
df.to_csv('MFDM_Web_Scrapping.csv', index=False, encoding='utf-8')
```
